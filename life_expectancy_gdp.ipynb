## Analysis: Life Expectancy  & GDP

### import libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load data

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
```

### check for missing data

```python
countries = df.Country.unique()
print(countries)
#years = df.Year.unique()
```

## Exploring the data

```python
sns.scatterplot(data = df, x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Country')
```

### Country by Country Comparison

```python
for i in countries:

    plt.subplot(2,2,1)
    sns.lineplot(data = df[df.Country == i], x = 'Year', y = 'GDP')
    plt.title(str(i) + '\n', fontdict = {'fontsize': 18,'horizontalalignment': 'left'})

    plt.subplot(2,2,3)
    sns.lineplot(data = df[df.Country == i], x = 'Year', y = 'Life expectancy at birth (years)')
    plt.ylabel('Life Expectancy')

    plt.subplot(1,2,2)
    sns.scatterplot(data = df[df.Country == i], x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Year')
    plt.legend(fontsize = 8)
    plt.ylabel('Life Expectancy')


    plt.subplots_adjust(wspace = .5)

    plt.show()
    plt.clf()

    print('\n\n')
```
